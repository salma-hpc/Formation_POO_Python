## 1. Attributs de classe

Pour bien comprendre la différence entre une donnée propre à un objet et une donnée partagée par toute la classe, on teste la création d'un **compteur d'objets**.

* **Le concept :** On crée une variable `num_instances` directement dans la classe. 
* **Le but :** Chaque fois qu'un nouvel objet est fabriqué, ce compteur augmente de 1.

```python
from compteur import ObjectCounter

# On fabrique 3 objets
obj1 = ObjectCounter()
obj2 = ObjectCounter()
obj3 = ObjectCounter()

# On vérifie que la classe a bien compté les 3
print(f"Nombre total d'objets créés : {ObjectCounter.num_instances}")
```

#### Le piège : Pourquoi ne pas utiliser `self` pour le compteur ?

Le tutoriel montre que si on utilise `self.num_instances` au lieu de `ObjectCounter.num_instances`, on crée une variable locale à l'objet au lieu de modifier le compteur global de la classe. 


```python
class ObjectCounterBad:
    num_instances = 0  # attribut de classe (initialisé à 0)

    def __init__(self):
        # on utilise self ici 
        self.num_instances += 1 
```

```python
# On crée 3 objets
bad1 = ObjectCounterBad()
bad2 = ObjectCounterBad()
bad3 = ObjectCounterBad()
```

```python
# On vérifie le compteur de la classse
print(f"Compteur de la classe : {ObjectCounterBad.num_instances}")

# On vérifie la valeur dans un objet précis
print(f"Valeur cachée dans l'objet bad1 : {bad1.num_instances}")
```

On voit bien que le compteur de la classe est resté à 0. 
L'utilisation de `self.num_instances` a créé un attribut propre à chaque objet au lieu de modifier l'attribut global. 

 Pour modifier un attribut de classe, il faut toujours utiliser le nom de la classe (`ObjectCounter.attribut`).

## 4. Attributs dynamiques de classe et d'instance
On peut ajouter des attributs à une classe vide au moment de l'exécution. C'est utile quand on ne connaît pas les données à l'avance (ex: lecture d'un fichier).

```python
# Définition d'une classe minimale "squelette"
class Record:
    """Hold a record of data.""" 
    pass # Mot-clé pour dire à Python : "Ne fais rien, la classe est vide pour l'instant"
```

```python
# Données simulées provenant d'un dictionnaire
john = {
    "name": "John Doe",
    "position": "Python Developer",
    "department": "Engineering",
    "salary": 80000,
    "hire_date": "2020-01-01",
    "is_manager": False,
}
```

```python
# Création de l'instance
john_record = Record()

# On parcourt le dictionnaire 'john' ligne par ligne
# 'field' recevra le nom du champ (ex: "name")
# 'value' recevra la donnée (ex: "John Doe")
for field, value in john.items():
    # On crée dynamiquement l'attribut dans l'objet john_record
    setattr(john_record, field, value)
```

```python
# Vérification des nouveaux attributs
print(john_record.name)
print(john_record.department)
print(john_record.__dict__)
```

##  6. Ajout dynamique d'une méthode d'initialisation
On peut même injecter une fonction dans une classe après sa définition pour qu'elle devienne une méthode (comme `__init__`).

```python
# Définition d'une fonction externe
def __init__(self, name, job):
    self.name = name
    self.job = job

# Injection de la fonction dans la classe User comme méthode d'initialisation
User.__init__ = __init__
```

```python
# Création d'un objet Linda en utilisant le constructeur injecté
linda = User("Linda Smith", "Team Lead")
print(linda.__dict__)
```

Comme le montre l'exemple précédent, il est possible de construire une classe Python entière de manière dynamique. Bien que cette fonctionnalité soit puissante, il convient de l'utiliser avec précaution car elle peut rendre le code difficile à comprendre et à analyser.

## 9. Optimisation mémoire avec `__slots__`
Par défaut, Python utilise un dictionnaire (`__dict__`) pour stocker les attributs, ce qui consomme beaucoup de mémoire. 
L'attribut `__slots__` permet de fixer la liste des attributs autorisés, supprimant le dictionnaire interne pour gagner de l'espace.

```python
class Point:
    __slots__ = ("x", "y")
    
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Création d'un point
p = Point(4, 8)
print(f"Point : x={p.x}, y={p.y}")
```

#### Test de la contrainte
Si on essaie d'ajouter un attribut `z` qui n'est pas dans les `__slots__`, Python lève une erreur `AttributeError`.

```python
p.z = 16
```

## 2. Les méthodes magiques : __str__ et __repr__ 
* `__str__` : Sert à donner une description jolie et lisible de l'objet (pour l'utilisateur).
* `__repr__` : Sert à donner une description technique et précise (pour le développeur).

```python
class Car:
    def __init__(self, make, model, year, color):
        self.make = make
        self.model = model
        self.year = year
        self.color = color

    # Méthode pour l'affichage "joli" (print)
    def __str__(self):
        return f"{self.make} {self.model}, {self.color} ({self.year})"

    # Méthode pour l'affichage technique (débogage)
    def __repr__(self):
        return (
            f"{type(self).__name__}"
            f'(make="{self.make}", '
            f'model="{self.model}", '
            f"year={self.year}, "
            f'color="{self.color}")'
        )

# test
ma_voiture = Car("Ford", "Mustang", 2022, "Noire")
```

```python
# Test du __str__ (appelé par print)
print(ma_voiture)

# Test du __repr__ (appelé quand on tape juste le nom de l'objet)
ma_voiture
```

# VII. Utilisation de l'héritage et construction de hiérarchies de classes

#### Pourquoi utiliser l'héritage ?
L'héritage permet de créer une relation hiérarchique entre une classe **Parente** (générale) et une classe **Enfant** (spécialisée). 

Les 3 grands piliers de l'héritage :
1. *Héritage d'implémentation* : Réutilisation du code existant pour éviter la duplication (DRY - *Don't Repeat Yourself*).
2. *Modularité* : Le code est mieux organisé, plus évolutif et plus facile à maintenir.
3. *Spécialisation d'interface* : Une classe enfant "est une" version plus précise de son parent (par exemple, un `Manager` *est un* `Employee`).

> **Concept clé :** Si une classe Enfant hérite d'une classe Parente, elle possède *automatiquement* tous les attributs et toutes les méthodes de celle-ci, sans avoir à les réécrire.

#### 1. Héritage simple
On parle d'*héritage simple* lorsqu'une classe enfant ne possède qu'une seule classe parente. C'est la forme la plus courante et la plus lisible de réutilisation de code.

**Syntaxe Python :**
Pour qu'une classe hérite d'une autre, on place le nom de la classe parente entre parenthèses juste après le nom de la classe enfant.

```python
class ClasseParente: # ClasseParente est la classe dont vous souhaitez hériter
    # Code de la classe de base
    pass

class ClasseEnfant(ClasseParente): # <--- Héritage simple ici
    # Code spécifique à la classe enfant
    pass
```

#### 2. Hiérarchies de classes (Arbres d'héritage) 

L'héritage ne se limite pas à deux niveaux. On peut créer des **hiérarchies complexes** où chaque sous-classe devient à son tour la base d'une classe encore plus spécialisée. 

C'est ce qu'on appelle la relation **"Est un"** (*Is-A*) :
* Un `Chien` **est un** `Mammifère`.
* Un `Mammifère` **est un** `Animal`.

Pour organiser cette nouvelle hiérarchie, nous utilisons un fichier séparé nommé `animals.py`. 

**Objectif du test :**
Nous allons instancier un objet `Shark` pour vérifier la **remontée d'héritage**. Un objet situé au bas de l'arbre doit pouvoir accéder aux informations de tous ses ancêtres :
1.  *Niveau 1 (Grand-parent - `Animal`)* : Attributs de base (`name`, `sex`, `habitat`).
2.  *Niveau 2 (Parent - `Fish`)* : Caractéristique biologique (`unique_feature`).
3.  *Niveau 3 (Enfant - `Shark`)* : Comportement spécifique (`swim`).

##### Au-delà de l'héritage : Les autres relations entre classes

L'héritage (relation **"Est un"**) n'est pas la seule façon de lier des classes. Pour concevoir des systèmes complexes, on utilise d'autres types d'associations :

| Relation | Type | Description | Exemple |
| :--- | :--- | :--- | :--- |
| **Héritage** | *Is-A* (Est un) | Une classe est une spécialisation d'une autre. | Un `Shark` est un `Animal`. |
| **Composition** | *Has-A* (Possède) | Relation forte : si le parent meurt, l'enfant aussi. | Un `Robot` a un `Bras`. |
| **Agrégation** | *Has-A* (Possède) | Relation faible : l'enfant peut survivre sans le parent. | Une `University` a un `Professor`. |
| **Association** | *Uses-A* (Utilise) | Une classe utilise simplement une autre. | Un `Student` suit un `Course`. |

Pour bien comprendre la différence avec l'héritage, créons un exemple de **Composition**. 
Ici, un `Robot` n'est pas un `Bras`, il **possède** un `Bras`. Si on détruit le robot, le bras (dans ce modèle) n'a plus d'utilité propre.

```python
# 1. Le composant (La pièce détachée)
# Cette classe est indépendante. Elle ne sait pas qu'elle appartient à un robot.
class RobotArm:
    def __init__(self, side):
        self.side = side

    def move(self):
        print(f"Le bras {self.side} effectue un mouvement.")

# 2. La classe principale (Le tout)
class Robot:
    def __init__(self, name):
        self.name = name
        #  LOGIQUE DE COMPOSITION 
        # Le Robot "possède" (Has-A) des instances de RobotArm.
        # Les bras sont créés en même temps que le robot.
        self.left_arm = RobotArm("gauche")
        self.right_arm = RobotArm("droit")

    def wave(self):
        # Le robot délègue le travail à ses composants
        print(f"{self.name} dit bonjour :")
        self.left_arm.move()
        self.right_arm.move()
```

```python
# Test
# On crée un robot, et automatiquement, deux bras sont créés à l'intérieur.
my_bot = Robot("R2D2")
my_bot.wave()
```

#### 5. Ordre de résolution de méthode (MRO)

L'héritage multiple peut créer une situation complexe appelée le **Problème du Diamant**. Cela arrive quand plusieurs parents proposent une version différente de la même méthode. 

*Comment Python choisit-il ?* Il utilise le **MRO** (*Method Resolution Order*). C'est un algorithme qui définit un chemin de recherche précis.

###### Le MRO (Method Resolution Order)

Pour éviter les conflits (le "Problème du Diamant"), Python suit un algorithme précis pour chercher les méthodes. Voici l'ordre de priorité en 5 étapes :

1.  *La classe actuelle* : Python cherche d'abord dans la classe qui appelle la méthode.
2.  *Les superclasses les plus à gauche* : Il regarde le premier parent listé dans la définition.
3.  *Les superclasses suivantes (de gauche à droite)* : Il parcourt les autres parents un par un.
4.  *Les parents des parents (ancêtres)* : Il remonte dans la généalogie de chaque branche.
5.  *La classe `object`* : C'est l'étape finale, car en Python, tout finit par hériter de la classe de base `object`.

*Remarque* : Il ne suffit pas de savoir que Python cherche chez les parents, il faut comprendre que l'ordre d'écriture est décisif :

* **La classe courante est toujours prioritaire** : Si tu réécris une méthode dans ta classe enfant, c'est celle-là qui sera utilisée, peu importe ce qu'il y a chez les parents.
* **La priorité de gauche à droite** : Si deux parents ont la même méthode, Python prendra celle du parent cité **en premier** dans la définition de la classe.

Pour illustrer le MRO, prenons l'exemple de hiérarchie de classes  (Module `mro.py`)

On pose la question : laquelle des versions sera appelée ?

```python
from mro import D
```

```python
D().method()
```

**Pourquoi c'est B qui gagne ?**
Bien que `B` et `C` possèdent toutes les deux la même méthode, Python choisit **B** car :
1. Il cherche d'abord dans la classe actuelle (`D`) -> Vide.
2. Il cherche ensuite dans les parents **de gauche à droite**.
3. Comme on a écrit `class D(B, C)`, **B** est à gauche et donc prioritaire.

C'est la preuve concrète que l'ordre d'écriture dans la parenthèse décide du comportement du programme.

##### *Attention au "MRO Error" (TypeError)*
Il est important de noter que Python ne peut pas toujours résoudre l'ordre des parents si la hiérarchie est trop complexe ou contradictoire (par exemple, si on essaie de créer des cycles d'héritage impossibles).

Dans ce cas, Python ne choisira pas au hasard : il bloquera l'exécution avec une **`TypeError`**. C'est une sécurité pour éviter que le programme n'ait un comportement imprévisible.

##### Vérification technique du MRO

Pour confirmer le chemin que nous venons d'analyser, Python nous permet de consulter directement l'ordre de résolution via l'attribut spécial `.__mro__`. C'est l'outil de diagnostic ultime pour le développeur.

```python

# Affichage de l'arbre généalogique technique
D.__mro__ 
```

```python
from pprint import pprint

# Ici on utilise la fonction pprint pour un affichage élegant 
pprint(D.__mro__)
```

#### **Les points forts :**
* *Réutilisabilité & Modularité* : On réutilise le code existant et on organise le projet en hiérarchies logiques.

# VIII. Alternatives à l'héritage : Composition et Délégation

L'héritage n'est pas l'unique solution pour réutiliser du code. Il existe d'autres techniques permettant de rendre le code plus robuste et flexible.

#### 1. La Composition : La relation "Possède un"
La *composition* permet de construire un objet complexe à partir de ses composants :
* *Assemblage* : On combine des objets qui fonctionnent comme des briques.
* *Flexibilité* : Elle est définie à l'exécution (*runtime*). On peut donc remplacer dynamiquement un composant par un autre du même type pour modifier le comportement de l'objet global.
* *Encapsulation* : L'objet composite utilise l'implémentation de ses composants sans forcément exposer toute leur interface.
> **Règle d'or : "Privilégier la composition à l'héritage"**
> Contrairement à l'héritage qui est rigide, la composition permet de concevoir des classes plus flexibles et évolutives.

*Exemple pratique* : Le robot industriel (`robot.py`)

Pour illustrer la composition, nous utilisons l'exemple d'un robot industriel. Contrairement à l'héritage, la classe `IndustrialRobot` ne "descend" d'aucune autre classe. Elle est **construite** en assemblant des composants indépendants : un corps (`Body`) et un bras (`Arm`).

C'est ici que la **délégation** entre en jeu : le robot reçoit un ordre (ex: `weld`), mais il délègue l'exécution réelle à son composant spécialisé (`arm`).

Voici comment utilisons  IndustrialRobot dans le code :

```python
from robot import IndustrialRobot
```

```python
# On crée l'objet robot (qui contient déjà son corps et son bras)
robot = IndustrialRobot()

# Le robot demande à son corps de pivoter vers la gauche
robot.rotate_body_left()

# Le robot demande à son bras de monter de 15 cm
robot.move_arm_up(15)

# Le robot ordonne à son bras de faire la soudure
robot.weld()
```

```python
# Le robot demande à son corps de pivoter de 20 degrés vers la droite
robot.rotate_body_right(20)

# Le robot demande à son bras de descendre de 5 cm
robot.move_arm_down(5)

# Le robot ordonne à son bras de recommencer la soudure
robot.weld()
```

##### . Bilan : Pourquoi la Composition est-elle plus robuste ?

L'exemple du robot met en lumière des avantages cruciaux par rapport à l'héritage :

* **Encapsulation et Sécurité** : Contrairement à l'héritage, la composition ne dévoile pas toute l'interface des composants. Un robot soudeur n'aura accès qu'à la fonction `.weld()` de son bras. Si on utilisait l'héritage, il pourrait hériter par erreur de fonctions dangereuses (comme la découpe ou le façonnage), risquant de provoquer des pannes.
* **Flexibilité à l'exécution** : On peut changer de bras dynamiquement (`robot.arm = NewArm()`) ou ajouter une méthode `.change_arm()`. Cela permet au robot de s'adapter à différentes technologies de soudage sans modifier sa structure.
* **Protection** : La composition protège les classes contre une utilisation non intentionnelle en n'exposant que les fonctionnalités strictement nécessaires.

> **En résumé** : La composition permet de construire des systèmes "à la carte" où chaque objet ne possède que les outils dont il a réellement besoin.

#### 2. La Délégation : La relation de "Dépendance"
La **délégation** modélise le fait qu'un objet dépend d'un autre pour accomplir une tâche spécifique. Au lieu de faire le travail lui-même, l'objet délègue la responsabilité à un autre objet spécialisé.


Pour illustrer ce concept de dépendance, nous allons créer un nouveau fichier nommé `stack.py`. 

L'objectif est de construire une pile (Stack) qui respecte la règle **LIFO** (Last-In, First-Out). Au lieu d'hériter d'une liste, ce qui serait dangereux car cela exposerait des méthodes interdites comme `.insert()` ou `.remove()`, nous allons **déléguer** la gestion des données à une liste interne privée.

```python
from stack import Stack
```

```python
stack = Stack([1, 2, 3]) # liste de 3 élements 
stack
```

```python
stack.push(4) # l'element 4 est rajouté à la fin de la liste
stack
```

```python
stack.pop() # le dernier élement sera retiré de la liste
stack
```

```python
stack.pop()
stack
```

```python
dir(stack)# # affiche la liste de toutes les méthodes et attributs accessibles dans l'objet stack
```

Comme le montre le résultat de `dir(stack)`, l'interface publique de notre classe est strictement limitée à `.push()` et `.pop()`. 

> **Retenir :** La délégation permet de transformer un objet "touche-à-tout" (la liste) en un outil spécialisé et sécurisé (la pile).

##### Pourquoi ne pas avoir choisi class Stack(list) ?

```python
class Stack(list):
     def push(self, item):
         self.append(item)
     def pop(self):
         return super().pop()
     def __repr__(self) -> str:
        return f"{type(self).__name__}({super().__repr__()})"
```

```python
stack = Stack()
dir(stack)
```

**Réponse** :la délégation est plus sûre que l'héritage pour créer une pile, car elle permet de cacher les fonctions inutiles ( `append`, `clear`, `remove`...)  et d'éviter les erreurs de manipulation

#### Automatiser la délégation : les méthodes magiques

Jusqu'ici, nous avons écrit chaque méthode de délégation manuellement (comme `push` et `pop`). Mais Python permet d'automatiser cela. 

Si on cherche un outil qu'un objet n'a pas, Python va regarder dans une méthode spéciale : `__getattr__()`. On peut s'en servir pour rediriger automatiquement la demande vers un autre objet spécialisé. 

Voici l'application de cette technique dans le fichier `serializer_delegation.py`

```python
from serializer_delegation import Employee
```

```python
# Création d'un employé (il a un nom, un âge et un salaire)
emp = Employee("Alice", 30, 50000)
```

```python
# On appelle .to_json() alors que cette méthode n'est pas dans Employee !
# C'est __getattr__ qui va attraper l'appel et l'envoyer au Serializer.
print("Test JSON :")
print(emp.to_json())
```

```python
# Pareil pour .to_pickle() !
# L'employé délègue encore une fois le travail au spécialiste.
print("\nTest Pickle (format binaire) :")
print(emp.to_pickle())
```

# IX. Création de classes de base abstraites (ABC) et d'interfaces

Pour mettre en pratique ce concept de "contrat" obligatoire pour nos formes géométriques, nous allons travailler dans le fichier : `shapes_abc.py`

L'idée ici est de définir un **"contrat obligatoire"** (une interface) que toutes nos classes devront respecter. 

**En résumé :**
1. *Un modèle commun* : On veut que chaque forme (Cercle, Carré, etc.) possède obligatoirement les méthodes `.get_area()` et `.get_perimeter()`.
2. *Le module `abc`* : C'est l'outil Python qui permet de créer ces classes "modèles".
3. *Pas d'objet "fantôme"* : On ne peut pas créer un objet directement depuis une classe abstraite. Elle ne sert que de base aux autres.

> **Note :** C'est une sécurité. Si une classe enfant oublie d'implémenter une méthode du modèle, Python bloquera la création de l'objet.

```python
from shapes_abc import Circle
```

```python
circle = Circle(100)
```

**Que s'est-il passé ?** J'ai essayé de créer un objet `Circle`, mais Python a renvoyé une erreur : 
`TypeError: Can't instantiate abstract class Circle with abstract method get_perimeter`.

**Pourquoi cette erreur ?**
* La classe parente `Shape` a défini un contrat : "Toute forme doit posséder `get_area` ET `get_perimeter`".
* Dans mon code, j'ai pas codé `get_perimeter`.
* **Résultat** : Python considère que `Circle` est encore "abstraite" (incomplète). Il interdit donc de créer l'objet pour éviter d'avoir un cercle qui ne saurait pas calculer son périmètre plus tard.

```python
circle.radius
```

**La solution :** Une fois que j'ai implémenté la méthode manquante `.get_perimeter()`, la classe `Circle` n'est plus considérée comme "abstraite". 

*Ce que cela prouve :*
* *Validation* : Python a vérifié que toutes les méthodes obligatoires définies dans `Shape` sont bien présentes. 
* *Utilisation* : L'objet `circle` peut maintenant être créé et utilisé normalement.

Désormais, n'importe quel développeur qui utilisera ma classe `Shape` sera obligé de suivre la même structure, ce qui garantit que toutes nos formes géométriques fonctionneront de la même manière.

En résumé : Les ABC servent à définir une "loi commune" pour toute une famille de classes, obligeant chaque enfant à coder ses propres versions des méthodes imposées.

# X. Déverrouiller le polymorphisme avec des interfaces communes

Le **Polymorphisme** permet d'utiliser des objets différents (comme des chaînes, des listes ou des tuples) de la même manière car ils partagent une **interface commune**.

**L'exemple des séquences :**
En Python, peu importe qu'on manipule du texte ou une liste, on peut utiliser les mêmes outils car ils respectent tous le "contrat" des séquences :
* **`len()`** : fonctionne car ils ont tous une méthode `.__len__()`.
* **Les boucles `for`** : fonctionnent car ils ont tous une méthode `.__iter__()`.
* **Les index ``** : fonctionnent car ils ont tous une méthode `.__getitem__()`.

C'est la magie du polymorphisme : on utilise l'objet sans se soucier de son type précis, tant qu'il sait répondre à nos commandes.

Considérons les exemples suivants, qui utilisent la fonction intégréelen() :

```python
message = "Hello!"
numbers = [1, 2, 3]
letters = ("A", "B", "C")
```

```python
len(message)
```

```python
len(numbers)
```

```python
len(letters)
```

Dans ces exemples, `len()` fonctionne sur trois objets différents (chaîne, liste, tuple) car ils partagent tous la même méthode `.__len__()`. C'est le principe du **polymorphisme** : on utilise différents objets de la même façon grâce à une interface commune.
